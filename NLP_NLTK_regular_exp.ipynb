```python
text = "hello !!! my name is sidhu  how are you doing mate ?  ok this is fine "
sentences = re.split(r'[.!?]',text)
sentences = [i.strip() for i in sentences if i.strip()]

print(sentences)
```

```python
count = 0 
for i,line in enumerate(sentences):
    line = re.sub(r'[^a-zA_Z0-9]',' ',line) #sub - substitute ^a-z , A_Z 0_9 other than this replace with
    sentences[i]=line.split()
    count+=len(sentences[i])
print(sentences)
print(count)
```
